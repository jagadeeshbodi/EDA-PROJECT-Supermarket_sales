# 1.Is your supermarket more popular with males or females?

```python
plt.figure(figsize=(14,8))
plt.style.use('fivethirtyeight')
ax=sns.countplot('Gender',data=df,palette='copper')
ax.set_xlabel(xlabel="Gender",fontsize=20)
ax.set_ylabel(ylabel="Gender count",fontsize=20)
ax.set_title(label="Gender count in supermarket",fontsize=22)
plt.show()
```

# 2.What does the customer rating look like and can you also comment on its skewness?

```python
sns.histplot(df['Rating'])
plt.axvline(x=np.mean(df['Rating']),c='red',ls='--',label='mean')
plt.axvline(x=np.percentile(df['Rating'],25),c='olive',ls='--',label='25th percentile:Q1')
plt.axvline(x=np.percentile(df['Rating'],75),c='brown',ls='--',label='75th percentile:Q3')
plt.legend()
```

```python
#Ans: the rating distribution looks uniform and it seems to be no skewness on the left or right side of the distribution.we can plot the mean and percentile to show this.
```
